# **LinearREgression**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data={
    "pizza" : [8,10,12],
    "price" : [10,13,16],
}
df = pd.DataFrame(data)
df
```

```python
#Feature and Target
x = df[["pizza"]]
y = df[["price"]]
```

```python

x_train, x_test, y_train, y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
new_price=[[30]]
predicted_sales=model.predict(new_price)
print("Predicted Sales:",predicted_sales[0])
```

# **Regression Matrix**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.metrics import r2_score
```

```python
data ={
    "study_hours" :[1,2,3,4,5,6,7,8],
    "marks" : [32,40,50,60,70,75,85,90]
}
df=pd.DataFrame(data)
df
```

```python
#Feature and Target
x = df[["study_hours"]]
y = df["marks"]
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
print("predicted marks:",y_pred)
```

MAE=(Mean Absolut Error)

```python
mae = mean_absolute_error(y,y_pred)
print("MAE:",mae)
```

MSE=(Mean Square Error)

```python
mse = mean_squared_error(y,y_pred)
print("MSE:",mse)
```

RMSE

```python
rmse=np.sqrt(mean_squared_error(y,y_pred))
print("RMSE:",rmse)
```

R2_score

```python
r2 = r2_score(y,y_pred)
print("R2_score:",r2)
```

# **Multicolarity**

```python
#step 1 create a sample dataset

import pandas as pd
data ={
    "hoise_size" : [800,1000,1200,1400,1600],
    "bedroom" : [2,3,3,4,5],
    "distance_city" : [15,12,10,8,6],
    "price" : [40,50,60,70,80]
}

df=pd.DataFrame(data)
df
```

```python
#step2 chek corelation matrix

corr_matrix=df.corr()
print(corr_matrix)
```

```python
#visulize co-relation

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr_matrix,annot=True,cmap="coolwarm" )

plt.show()
```

```python
#cal VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

x=df[["hoise_size","bedroom","distance_city"]]

x["Intercept"] = 1

vif_data=pd.DataFrame()
vif_data["Feature"] =x.columns
vif_data['VIF']=[variance_inflation_factor(x.values,i)
for i in range(x.shape[1])
]
print(vif_data   )

```
