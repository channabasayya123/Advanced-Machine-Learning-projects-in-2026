```python

import pandas as pd
df=pd.read_csv("data_cleaning_dataset_1.csv")
#df.head() # reading first 5 rows
print(df)
```

```python
print("dataset shape:",df.shape)
```

```python
df.duplicated()
```

```python
#remove dupictae values
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
# handling missing values
# numerical columns--> fill with median value
num_cols=["Age","Marks","Attendance_Percent"]
#with loop
for col in num_cols:
  df[col]=df[col].fillna(df[col].median())
print(df)
```

```python
# without for loop
num_cols=["Age","Marks","Attendance_Percent"]
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
print(df)
```

```python
#catgorical columns
cat_cols=["Gender","Department","City","Fee_Paid"]

#with loop
for col in cat_cols:
  df[col]=df[col].fillna(df[col].mode()[0])
print(df)
```

```python
df.isnull().sum()
```

```python
#Email ---> fill missing with "Unknown"

df['Email']=df['Email'].fillna("Unknown")
print(df)
```

```python
# Gender Standardization
df['Gender']=df['Gender'].replace({
    "M":"Male",
    "male":"Male",
    "F":"Female",
    "female":"Female"
})
print(df)
```

```python
# Fee_paid Standardization
df['Fee_Paid']=df['Fee_Paid'].replace({
    "N":"No",
    "Y":"Yes",
    "yes":"Yes",
    "no":"No"
})
print(df)
```

```python
#Binary encoding
df['Scholarship']=df['Scholarship'].map({"Yes":1,"No":0})
df['Hostel']=df['Hostel'].map({"Yes":1,"No":0})
df['Fee_Paid']=df['Fee_Paid'].map({"Yes":1,"N0":0})
print(df)
```

```python
#One Hot Encoding
df=pd.get_dummies(
    df,
    columns=["Blood_Group"],
    drop_first=True
    )
print(df)
```

```python
df.to_csv("data_cleaning_dataset_1.csv", index=False)
from google.colab import files
files.download("data_cleaning_dataset_1.csv")
```
