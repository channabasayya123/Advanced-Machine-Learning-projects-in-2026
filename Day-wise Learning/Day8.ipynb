# **Decision Tree**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data={
    'hours_studies' : [1,2,3,4,5,6,7,8],
    'attendence' : [40,50,60,70,80,90,85,95],
    'pass' : [0,0,0,1,1,1,1,1]
}
df=pd.DataFrame(data)

x=df[['hours_studies','attendence']]
y=df['pass']
```

```python

x_train, x_test, y_train, y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
model=DecisionTreeClassifier(criterion='gini',max_depth=3)
model.fit(x_train,y_train)
```

```python
y_pred = (x_test)
y_pred=model.predict(y_pred)
print("Predicted:",y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plot_tree(
    model,
    feature_names=['hours_studies','attendence'],
    class_names=['Fail','Pass'],
    filled=True
)
plt.show()
```

# **Random Forest**

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_iris()
x=data.data
y=data.target

```

```python
x_tarin,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
rf=RandomForestClassifier(
  n_estimators=100,
  bootstrap=True,
  random_state=42
 )
rf.fit(x_tarin,y_train)
```

```python
y_pred = rf.predict(x_test)
print("predicted:",y_pred)

```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plot_tree(
    rf.estimators_[0],
    feature_names=data.feature_names,
    class_names=data.target_names,
    filled=True,
    rounded=True
)
plt.title('One Decision Tree Random Forest(Iris Dataset)')
plt.show()
```

```python
#get feature imp
importance=rf.feature_importances_
features = data.feature_names
```

```python
import matplotlib.pyplot as plt


plt.figure()
plt.bar(features,importance)
plt.ylabel("Importances score")
plt.xlabel("features imp_random forest")
plt.xticks(rotation=45)
plt.show()
```
